```python
#remove duplicates from list below
my_list=[1,2,3,3,2,4,5,6]

print(set(my_list))
```

```python
#find common elements in 2 sets
a=[1,2,3,4]
b=[3,4,5,6]

common = list(set (a) & set(b))
print(common)
```

```python
# get only unique elemnts from two lists
a =[1,2,3,4]
b= [3,4,5,6]

unique =list (set(a) ^ set (b))
print (unique)
```

```python
#detect duplicates in below list
my_list=[1,2,3,4,4,5,5]
my_set = set(my_list)
if(len(my_list)!= len(my_set)):
    print('has duplicates')
```
